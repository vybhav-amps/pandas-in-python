# Cleaning the Data:

```python
import pandas as pd
```

```python
df=pd.read_csv('dirtydata.csv')
df
```

The data set contains some empty cells. (NaN values)

The data set contains wrong format. (integer instred of string)

The data set contains wrong data.  (data limit extend)

The data set contains duplicates.   (repeations of rows and columns)

### Empty cells 

```python
new_df = df.dropna()  #removes rows
new_df
```

```python
#replace null values with values

new_df=df.fillna(100)
new_df
```

```python
#Replace Only For a Specified Columns

df=pd.read_csv('dirtydata.csv')
df["Calories"].fillna(100, inplace = True)
df
```

#### Replace Using Mean, Median, or Mode

```python
#Calculate the MEAN, and replace any empty values with it

df=pd.read_csv('dirtydata.csv')

x = df["Calories"].mean()

df["Calories"].fillna(x, inplace = True)
df
```

```python
#Calculate the MEDIAN, and replace any empty values with it

df=pd.read_csv('dirtydata.csv')

x = df["Calories"].median()

df["Calories"].fillna(x, inplace = True)
df
```

```python
#Calculate the MODE, and replace any empty values with it

df=pd.read_csv('dirtydata.csv')

x = df["Calories"].mode()[0]

df["Calories"].fillna(x, inplace = True)
df
```

### Wrong format

```python
#convert all cells in the 'Date' column into dates

df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
#Remove rows with a NULL value in the "Date" column

df.dropna(subset=['Date'], inplace = True)
df
```

### Wrong Data

```python
for x in df.index:
   if df.loc[x, "Duration"] > 60:
    df.loc[x, "Duration"] = 45
df
```

### Duplicates

```python
df.duplicated()
```

```python
df.drop_duplicates(inplace = True)
df
```
